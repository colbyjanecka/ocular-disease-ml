```python
import os
import cv2
import matplotlib.pyplot as plt

def cropImageSet(name):

    inputDataPath = "ocular_disease_fundus/ODIR-5K/ODIR-5K/Actual Training Images/" + name
    outputDataPath = inputDataPath + "-cropped"
    if(not os.path.isdir(outputDataPath)): os.mkdir(outputDataPath) 


    def crop_image_only_outside(img,tol=1):
        # img is 2D or 3D image data
        # tol  is tolerance
        mask = img>tol
        if img.ndim==3:
            mask = mask.all(2)
        m,n = mask.shape
        mask0,mask1 = mask.any(0),mask.any(1)
        col_start,col_end = mask0.argmax(),n-mask0[::-1].argmax()
        row_start,row_end = mask1.argmax(),m-mask1[::-1].argmax()
        return img[row_start:row_end,col_start:col_end]



    for filename in os.listdir(inputDataPath):

        if filename.endswith(".jpg"):

            thisImg = cv2.imread(inputDataPath + "/" + filename)

            print("Cropping " + filename)

            croppedImg = crop_image_only_outside(thisImg)

            if(cv2.imwrite(outputDataPath + "/" + filename, croppedImg)):
                print("Saved " + filename)


```

```python
diseases = ["amd", "cataracts", "diabetes", "glaucoma", "hypertension", "myopia", "normal"]

for disease in diseases:
    print("Croppping set for " + disease)
    cropImageSet(disease)
```
